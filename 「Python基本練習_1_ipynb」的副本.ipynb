```python
#Python範例
# while 迴圈
# 計算等差級數總和

sum = 0   # 總和變數
i = 1   # 索引
while i <= 10 :
    print('i=',i)
    sum = sum+i
    i = i+1

print('總和=',sum)
```

```python
#Python範例
# while 迴圈
# 計算等比級數總和

sum = 0   # 總和變數
x = 1   #
while x > 0.001 :
    print('x=',x)
    sum = sum+x
    x=x/2

print('總和=',sum)
```

```python
#Python範例
# 用 import 匯入函式庫
# 繪圖2：xy 函數繪圖
import numpy as np #匯入數值計算函式庫
import matplotlib.pyplot as plt #匯入繪圖函式庫

#設定繪圖範圍
x1=-20
x2=20
y1=-4
y2=4
plt.axis([x1,x2,y1,y2])
plt.axis('on') #標出xy軸
plt.grid(True) #要畫格線

#產生一組x數列
x = np.arange(x1, x2, 0.1)

#產生對應的y值
y = np.sin(x)
yp = np.cos(x)

plt.scatter(x,y,s=1,color='red') #畫點
plt.plot(x,yp,linewidth=3,color='blue') #畫線

plt.show() #秀出圖形
```

```python
#Python範例
# 用 import 匯入函式庫
# 繪圖4：玫瑰線
import numpy as np #匯入數值計算函式庫
import matplotlib.pyplot as plt #匯入繪圖函式庫

#設定繪圖範圍
x1=-1.5
x2=1.5
y1=-1.5
y2=1.5

###################################
#設定玫瑰線參數，請改變這兩個值
###################################
n=16.0
d=15.0
###################################


plt.axis([x1,x2,y1,y2]) #設繪圖區範圍
plt.axis('off') #不顯示xy軸
plt.grid(False) #不要畫格線

#用極座標畫玫瑰線

theta = np.arange(0,100*np.pi,0.1) #角度由0到100pi
r = np.cos(n/d*theta) #計算角度對應的半徑
x = r*np.cos(theta) #把極座標(r,theta)換成(x,y)座標
y = r*np.sin(theta) #把極座標(r,theta)換成(x,y)座標

plt.plot(x,y,linewidth=2,color='red') #畫線

plt.show() #秀出來
```
